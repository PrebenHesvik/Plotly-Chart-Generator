```python
import pandas as pd
import numpy as np
import os
import sys

module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)

from plotly_chart_generator.subplots.pie_subplots import pie_subplots
from plotly_chart_generator.subplots.scatter_subplots import scatter_subplots
from plotly_chart_generator.chart_styles import chart_styles
```

# Pie subplots

```python
# Create data
index = ['Production', 'Assembly', 'Design']
values = {'Unit 1': [55, 35, 10], 
         'Unit 2': [50, 30, 20],
         'Unit 3': [45, 45, 10],
         'Unit 4': [40, 45, 15],
         }

data = pd.DataFrame(data=values, index=index)

# Create chart
layout = chart_styles(title='Pie Plots', height=600, width=600)

pie_subplots(data=data.to_dict(), rows=2, cols=2,
                   titles=[*data.to_dict().keys()], layout=layout,
                   hoverinfo='label+name', textinfo='value')
```

## Scatter subplots

```python
center = (50, 60)
distance = 5
size = 100

data_dict = {}
for letter in 'ABCD':
    #center = np.random.randint(low=40, high=50, size=1), np.random.randint(low=40, high=50, size=1)
    x = np.random.uniform(center[0], center[1] + distance, size=(size,))
    y = np.random.normal(center[1], distance, size=(size,))
    names = [str((i, ii)) for i, ii in zip(x, y)]
    data_dict.update({letter: {
        'x':x,
        'y':y,
        'names': names
    }})
        
layout = chart_styles(title='Spend/antall faktura fordelt på leverandør/lokasjon',
                               height=800, width=800)

scatter_subplots(data_dict, rows=2, cols=2, titles=list(data_dict.keys()), layout=layout)
```
